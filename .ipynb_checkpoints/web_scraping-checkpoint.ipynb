```python
import pandas as pd
from urllib.parse import urlparse
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import OneHotEncoder
```

```python
# Upload
data_path = 'centr_seek-training_data.xlsx'
data = pd.read_excel(data_path)

```

## Preprocessing and checking data if parsable creating a parsed 

```python
# dropping all null values in the dataset
examples = data['Example'].dropna().reset_index(drop=True)
non_examples = data['Non-Example'].dropna().reset_index(drop=True)

# Merging data frame to display the below
combined_data = pd.concat([
    pd.DataFrame({'URL': examples, 'Label': 1}),
    pd.DataFrame({'URL': non_examples, 'Label': 0})
], ignore_index=True)


print(combined_data)
```

## Feature extraction from the URL for each Community and Non community based research

### URL feature extraction  as per below criteria:
1. Domain --> Domain part
2. path_length --> length of the URL path string
3. path_depth --> depth of the URL
4. news, community and  research in the link

## example --> 
URL: https://news.iu.edu/live/news/26186-indiana-professor-promotes-community-research

(
    'news.iu.edu',    <-- domain

    6,                <-- url_depth
 
    59,               <-- url_length
  
    1,                <-- url_news
   
    1,                <-- url_community
    
    1                 <-- url_research
)


```python
# approach of URL Parsing

def extract_url_features(url):
    parsed_url = urlparse(url)
    domain = parsed_url.netloc
    url_link = parsed_url.path
    url_depth = url_link.count('/')
    url_length = len(url_link)
    url_news = 1 if 'news' in url_link else 0
    url_community = 1 if 'community' in url_link else 0
    url_research = 1 if 'research' in url_link else 0
    return domain, url_depth, url_length, url_news, url_community, url_research
```

## One Hot Encoding

A --> 	[1, 0, 0] <br>
B --> 	[0, 1, 0] <br>
C --> 	[0, 0, 1]

```python
# feature extraction 
# convert the extracted data to into list
# combine features extracted data and URL extraction (combined data)
features = combined_data['URL'].apply(lambda x: extract_url_features(x))
features_df = pd.DataFrame(features.tolist(), columns=['Domain', 'Path Depth', 'Path Length', 'Has News', 'Has Community', 'Has Research'])
enhanced_data = pd.concat([combined_data, features_df], axis=1)

# One hot encoder for conversion of categorical data into numerical value
# fit transform (sparse matrix)
# To array converts to dense matrix
encoder = OneHotEncoder()
encoded_domains = encoder.fit_transform(enhanced_data[['Domain']]).toarray()  #sparse matrix to dense matrix this resolves the error
encoded_domains_df = pd.DataFrame(encoded_domains, columns=encoder.get_feature_names_out(['Domain']))

# Combine the encoded domain features
final_data = pd.concat([enhanced_data.drop(['Domain', 'URL'], axis=1), encoded_domains_df], axis=1)

# Split the data into training and testing sets
# 20% for split data 
# random number to ensure reproducibility
X_train, X_test, y_train, y_test = train_test_split(final_data.drop('Label', axis=1), final_data['Label'], test_size=0.2, random_state=42)


# output for exmaple data
print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])
print("Example of training data (features):\n", X_train.head())
print("Example of training data (labels):\n", y_train.head())
print("Example of testing data (features):\n", X_test.head())
print("Example of testing data (labels):\n", y_test.head())


# Initialize Random Forest classifier
# n_estimators sets the value for number of tress to 100 
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Prediction on the test set and evaluate the classifier
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```

## Function to extract a feature and test a link

```python
# This fuinction passes in URL
# performs one hot encoding
# model is based on the trained model rf classifier
def prepare_and_predict_url(url, encoder, model):
    
    domain, path_depth, path_length, has_news, has_community, has_research = extract_url_features(url)
    
    # URL extraction is performed below
    features = pd.DataFrame({
        'Path Depth': [path_depth],
        'Path Length': [path_length],
        'Has News': [has_news],
        'Has Community': [has_community],
        'Has Research': [has_research]
    })
    
    # One-hot encode for the domain
    encoded_domain = encoder.transform([[domain]]).toarray()
    encoded_domain_df = pd.DataFrame(encoded_domain, columns=encoder.get_feature_names_out(['Domain']))
    
    # Combine features to same df
    features = pd.concat([features, encoded_domain_df], axis=1)
    prediction = model.predict(features)
    return "Community-based research" if prediction[0] == 1 else "Not community-based research"

# Testing by calling the function
custom_url = "https://news.iu.edu/feature/10-a-new-era-in-indianapolis"
result = prepare_and_predict_url(custom_url, encoder, rf_classifier)
print("Prediction for the URL:", result)
```

## Scraping the data again for visualization


creating a function for that requests a powerful library for HTTP requests <br>
Here I'm sending a get request for fetching the data <br>
Exception handling code <br>
if status code == 200 then perform BeautifulSoup lib to return the response <br>
else throw the error <br>

```python
import requests
from bs4 import BeautifulSoup

def fetch_text_from_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            return soup.get_text(separator=' ', strip=True)
        else:
            return ''
    except Exception as e:
        return f'Error: {e}'

```

```python
file_path = 'centr_seek-training_data.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Fetch text from the URLs for community engaged (Example)
# starts by dropping null values in the dataset
example_urls = df['Example'].dropna()
example_texts = [fetch_text_from_url(url) for url in example_urls]
example_text_combined = ' '.join(example_texts)

# same thing as non example dataset (non community engaged)
non_example_urls = df['Non-Example'].dropna()
non_example_texts = [fetch_text_from_url(url) for url in non_example_urls]
non_example_text_combined = ' '.join(non_example_texts)

print("Scraping complete for Example and Non-Example URLs.")
```

## Word cloud gives a idea about the intesity of the words

```python
# import libs STOP words and wordcloud
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to preprocess text for scraped data
# condition to check if the data is in STOPWORDS or not
def preprocess_text(text):
    words = text.split()
    words = [word for word in words if word.lower() not in ENGLISH_STOP_WORDS]
    return ' '.join(words)

# Preprocess the combined texts by applying above function
example_text_preprocessed = preprocess_text(example_text_combined)
non_example_text_preprocessed = preprocess_text(non_example_text_combined)

# Generate and display word clouds
def generate_word_cloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=20)
    plt.axis('off')
    plt.show()

generate_word_cloud(example_text_preprocessed, 'Word Cloud for Example (Community Engaged)')
generate_word_cloud(non_example_text_preprocessed, 'Word Cloud for Non-Example (Not Community Engaged)')
```

## N-grams frequency <-- to point out the intensity of words 

```python
import nltk
from nltk.util import ngrams
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

# nltk.download('punkt')
# nltk.download('stopwords')

# Stopword list
stop_words = set(stopwords.words('english'))

# Firstly tokenizing into individual words
# checking the condition for stop words and isalpha()
def preprocess_text(text):
    words = nltk.word_tokenize(text)
    words = [word for word in words if word.lower() not in stop_words and word.isalpha()]
    return words

# Generates n-grams
def generate_ngrams(words, n):
    n_grams = ngrams(words, n)
    return [' '.join(grams) for grams in n_grams]

# Top n frequent n-grams or words
def get_top_n_frequencies(items, top_n=10):
    item_freq = Counter(items)
    return item_freq.most_common(top_n)

# Visualize top n-grams or words 
def plot_top_n_frequencies(item_freq, title):
    items, freq = zip(*item_freq)
    plt.figure(figsize=(10, 5))
    plt.barh(items, freq, color='skyblue')
    plt.xlabel('Frequency')
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.show()

# Combining texts by preprocessing it
example_words_preprocessed = preprocess_text(example_text_combined)
non_example_words_preprocessed = preprocess_text(non_example_text_combined)


top_10_example_words = get_top_n_frequencies(example_words_preprocessed, 10)
plot_top_n_frequencies(top_10_example_words, 'Top 10 Words for Example (Community Engaged)')

top_10_non_example_words = get_top_n_frequencies(non_example_words_preprocessed, 10)
plot_top_n_frequencies(top_10_non_example_words, 'Top 10 Words for Non-Example (Not Community Engaged)')

top_10_example_bigrams = get_top_n_frequencies(generate_ngrams(example_words_preprocessed, 2), 10)
plot_top_n_frequencies(top_10_example_bigrams, 'Top 10 Bigrams for Example (Community Engaged)')

top_10_example_trigrams = get_top_n_frequencies(generate_ngrams(example_words_preprocessed, 3), 10)
plot_top_n_frequencies(top_10_example_trigrams, 'Top 10 Trigrams for Example (Community Engaged)')

top_10_non_example_bigrams = get_top_n_frequencies(generate_ngrams(non_example_words_preprocessed, 2), 10)
plot_top_n_frequencies(top_10_non_example_bigrams, 'Top 10 Bigrams for Non-Example (Not Community Engaged)')

top_10_non_example_trigrams = get_top_n_frequencies(generate_ngrams(non_example_words_preprocessed, 3), 10)
plot_top_n_frequencies(top_10_non_example_trigrams, 'Top 10 Trigrams for Non-Example (Not Community Engaged)')
```
